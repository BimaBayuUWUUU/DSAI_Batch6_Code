## **Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.**
* Normal: 18.5-24.9
* Overweight: 25.0-29.9
* Obesity: 30.0-34.9
* Underweight: < 18.5


```python
# Buat status BMI untuk setiap siswa dan tambahkan kolom baru
def get_status_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi <= 29.9:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.9:
        return 'Obesity'
    else:
        return 'Tidak Ada Interval Yang Sesuai'

df_17['Status BMI'] = df_17['BMI'].apply(get_status_bmi)
```

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

```python
def calculate_investment(capital, tenure):
    # Cek apakah modal investasi dan jangka waktu investasi sesuai ketentuan
    if capital < 0 or capital > 1000000000 or tenure < 1 or tenure > 25:
        if capital < 0 or capital > 1000000000:
            print("Error: Modal investasi harus dalam rentang 0 hingga 1.000.000.000.")
        if tenure < 1 or tenure > 25:
            print("Error: Jangka waktu investasi harus dalam rentang 1 hingga 25 tahun.")
        return

    # Hitung keuntungan setiap tahun
    total_investment = capital
    for year in range(1, tenure + 1):
        profit = total_investment * 0.25
        total_investment += profit
        print(f"Tahun ke- {year}, total investasi Anda: {total_investment}")

# Menerima input dari pengguna
try:
    print("==========================================================")
    print("----------------Investasi Jamin-Sugih Co------------------")
    print("==========================================================")
    capital = int(input("Modal Investasi               : "))
    tenure = int(input("Jangka Waktu Investasi (tahun): "))
    print("==========================================================")
    print("----------------------Hasil Inventasi---------------------")
    print("==========================================================")
    calculate_investment(capital, tenure)
except ValueError:
    print("Error: Masukkan harus berupa bilangan bulat.")
```
