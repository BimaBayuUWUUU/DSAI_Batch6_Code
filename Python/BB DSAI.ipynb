```python
# Age - mean
df_miss_values['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df_miss_values['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df_miss_values['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```

## Duplicate Value

```python
df_miss_values.duplicated().sum()
```

```python
df_dup_fixed = df_miss_values.drop_duplicates()
```

## Outlier

```python
kolom_kolom = ['Age', 'Fare']

batas_kolom = {}
```

```python
for kolom in kolom_kolom:
    Q1 = df_dup_fixed[kolom].quantile(0.25)
    Q3 = df_dup_fixed[kolom].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    batas_kolom[kolom] = (lower_bound, upper_bound)

outliers = {}
for kolom in kolom_kolom:
    lower_bound, upper_bound = batas_kolom[kolom]
    outliers[kolom] = df_dup_fixed[(df_dup_fixed[kolom] < lower_bound) | (df_dup_fixed[kolom] > upper_bound)]
```

```python
for kolom, (lower_bound, upper_bound) in batas_kolom.items():
    print(f"Kolom '{kolom}':")
    print(f"  Lower Bound: {lower_bound}")
    print(f"  Upper Bound: {upper_bound}")
    print()
```

```python
print(f"Outliers pada kolom 'Age':")
outliers['Age']
```

```python
print(f"Outliers pada kolom 'Fare':")
outliers['Fare']
```

```python
# Inisialisasi DataFrame baru untuk menyimpan data selain outlier
df_no_outliers = df_dup_fixed.copy()

# Loop melalui setiap kolom dan hapus baris yang merupakan outlier
for kolom in kolom_kolom:
    lower_bound, upper_bound = batas_kolom[kolom]
    df_no_outliers = df_no_outliers[~((df_no_outliers[kolom] < lower_bound) | (df_no_outliers[kolom] > upper_bound))].reset_index(drop=True)

# Cetak DataFrame baru yang hanya berisi data selain outlier
print("DataFrame baru tanpa outlier:")
df_no_outliers
```

## Imbalance

```python
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
```

```python
df = df_no_outliers.copy()

# Memisahkan fitur dan target
X = df.drop('Survived', axis=1)
y = df['Survived']

# Menampilkan jumlah X dan y sebelum SMOTE
print("Jumlah X sebelum SMOTE:", X.shape)
print("Jumlah y sebelum SMOTE:", y.shape)
```

```python

# Menggunakan SMOTE untuk oversampling
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Menampilkan jumlah X dan y setelah SMOTE
print("Jumlah X setelah SMOTE:", X_resampled.shape)
print("Jumlah y setelah SMOTE:", y_resampled.shape)
```

```python
# Visualisasi dengan bar plot
labels = ['Sebelum SMOTE', 'Setelah SMOTE']
jumlah_X = [X.shape[0], X_resampled.shape[0]]
jumlah_y = [y.shape[0], y_resampled.shape[0]]

plt.figure(figsize=(8, 6))
plt.bar(labels, jumlah_X, color='blue', alpha=0.6, label='Jumlah X')
plt.bar(labels, jumlah_y, color='orange', alpha=0.6, label='Jumlah y')
plt.title('Jumlah Sampel Sebelum dan Setelah SMOTE')
plt.ylabel('Jumlah Sampel')
plt.xlabel('SMOTE')
plt.legend()
plt.show()
```

```python

# Mengganti dataframe asli dengan dataframe yang sudah dioversampling
df_oversampled = pd.concat([X_resampled, y_resampled], axis=1)

# Menampilkan dataframe yang sudah dioversampling
print("\nDataFrame yang sudah dioversampling:")
print(df_oversampled)
```
