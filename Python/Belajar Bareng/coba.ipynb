Model yang kami pilih adalah Decision Tree Classifier. Model ini dipilih karena model ini memiliki akurasi yang cukup baik dan memiliki Learning Curve yang stabil. Selain itu, model ini juga memiliki waktu komputasi yang relatif cepat jika dibandingkan 2 model lainnya, yaitu Adaptive Boosting dan Extreme Gradient Boosting. Fitur yang kami gunakan adalah fitur yang sudah kami pilih sebelumnya, yaitu fitur yang seacara rasionalitas dapat mempengaruhi target. Hal ini karena korelasi antar fitur dengan target yang rendah. Evaluation metric yang kami gunakan adalah Accuracy, Precision, Recall, F1 Score, dan Learning Curve. Kami menggunakan metric ini karena kami ingin mengetahui seberapa baik model kami dalam melakukan klasifikasi. Dan dibuktikan jika nilai akurasi, precision, recall, dan f1 score saling linear.

Model Awal kami adalah Adaboost karena jika dilihat secara akurasi dan LC memiliki performa yang paling baik. LC pada model Decision Tree Classifier juga lebih stabil karena Training Score dan Cross Validation Score linear. Garis Validation Score juga mengikuti garis Training Score. Gap antara Training Score dan Cross Validation Score juga tidak terlalu jauh berkisar 0.2, Sedangkan gap Training-Validation mdoel Adabost mencapai 0.5.
```python
# Clustering menggunakan K-Means
kmeans = KMeans(n_clusters=10, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

print(df.head())
```
