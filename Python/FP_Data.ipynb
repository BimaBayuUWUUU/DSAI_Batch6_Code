```python
# Path file CSV
path_file1 = '../Dataset/data_FP/detail_pekerjaan/skill_pekerjaan.csv'

# Membaca file CSV ke dalam DataFrame
df1 = pd.read_csv(path_file1)

# Menggunakan value_counts() untuk mendapatkan nilai unik dan frekuensinya
unique_counts1 = df1['id_pekerjaan'].value_counts()

# Menampilkan nilai unik dan frekuensinya
print(unique_counts1)
```

```python
# Mengelompokkan berdasarkan 'id_pekerjaan' dan mengumpulkan 'singkatan_kemampuan' sebagai list
df_grouped1 = df1.groupby('id_pekerjaan')['singkatan_kemampuan'].apply(list).reset_index()

# Membuat kolom baru untuk industri
for i in range(3):
    df_grouped1[f'skill_{i+1}'] = df_grouped1['singkatan_kemampuan'].apply(lambda x: x[i] if i < len(x) else 'NOT')

# Menghapus kolom 'singkatan_kemampuan' asli
df_grouped1.drop('singkatan_kemampuan', axis=1, inplace=True)

# Menampilkan DataFrame akhir
print(df_grouped1)
```

```python
# Path file CSV
path_file2 = '../Dataset/data_FP/detail_pekerjaan/benefit.csv'

# Membaca file CSV ke dalam DataFrame
df2 = pd.read_csv(path_file2)

# Membuat DataFrame baru yang berisi jumlah kejadian 'benefit' == 1 untuk setiap nilai unik dalam 'id_pekerjaan'
benefit_counts = df2[df2['benefit'] == 1].groupby('id_pekerjaan')['benefit'].count()

# Membuat DataFrame baru dari benefit_counts dengan mereset index
df_benefit_counts = pd.DataFrame(benefit_counts).reset_index()

# Menampilkan hasil
print(df_benefit_counts)
```

```python
# Path file CSV
path_file_kar = '../Dataset/data_FP/detail_perusahaan/jumlah_karyawan.csv'

# Membaca file CSV ke dalam DataFrame
df_jumkar = pd.read_csv(path_file_kar)
```

```python
# Menyortir DataFrame berdasarkan id_perusahaan dan waktu_dicatat
df_sorted = df_jumkar.sort_values(by=['id_perusahaan', 'waktu_dicatat'])

# Menghapus duplikat berdasarkan id_perusahaan dengan mempertahankan yang terakhir (nilai waktu_dicatat yang terbesar)
df_filtered_jumkar = df_sorted.drop_duplicates(subset='id_perusahaan', keep='last')
```

```python
# Lakukan left outer join untuk ketiga DataFrame
result = pd.merge(pd.merge(df_main, df, on='id_pekerjaan', how='left'), df_grouped1, on='id_pekerjaan', how='left')
```
