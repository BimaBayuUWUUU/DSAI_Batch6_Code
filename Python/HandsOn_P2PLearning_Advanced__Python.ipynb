# **Fungsi dan Argumen**

## **Fungsi Bersarang**

```python
tinggi_segitiga = [5, 8, 10, 6, 7]
alas_segitiga = [3, 6, 4, 9, 8]
tinggi_prisma = [12, 15, 20, 33, 10]
```

```python
def hitung_luas_segitiga(tinggi, alas):
    luas = 0.5 * tinggi * alas
    return luas
```

```python
def hitung_volume_prisma_segitiga(tinggi_prisma, tinggi_segitiga, alas_segitiga):
    luas_alas = hitung_luas_segitiga(tinggi_segitiga, alas_segitiga)
    volume = luas_alas * tinggi_prisma
    return volume
```

```python
volume_prisma = []

# Menghitung volume prisma segitiga untuk setiap segitiga dalam list
for i in range(len(tinggi_segitiga)):
    volume = hitung_volume_prisma_segitiga(tinggi_prisma[i], tinggi_segitiga[i], alas_segitiga[i])
    volume_prisma.append(volume)
```

## **Arbitary**

```python
def gabung(*args, **kwargs):
    hasil = ""
    for arg in args:
        hasil += arg
    for key, value in kwargs.items():
        hasil += f" {key} {value}"
    return hasil
```

# **Data Frame**

## **Import Library**

```python
import numpy as np
import pandas as pd
```

## **Load Dataset**

```python
file_path = "https://raw.githubusercontent.com/BimaBayuUWUUU/DSAI_Batch6_Code/main/Dataset/diabetes.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
df.head(5)
```

## **Sorting**

```python
# Mengurutkan DataFrame berdasarkan BMI
sorted_df = df.sort_values(by='BMI', ascending = True, kind='mergesort')
```

```python
sorted_df.head(5)
```

## **Filtering**

```python
# Melakukan filtering untuk beberapa kolom dengan nilai 0
filtered_df = df[(df['BMI'] > 0.0) &
                (df['SkinThickness'] > 0) &
                 (df['BloodPressure'] > 0) &
                 (df['Insulin'] > 0) &
                 (df['Glucose'] > 0)]
```

```python
filtered_df.head(5)
```

## **Grouping dan Aggregating**

```python
# Aggregasi untuk menghitung rata-rata kadar Insulin, Glukosa, dan Resiko Diabetes serta modus nilai BMI untuk pasien dengan dan tanpa diabetes
aggregated_df = filtered_df.groupby('Outcome').agg({'Insulin': 'mean',
                                                    'Glucose': 'mean',
                                                    'BMI': lambda x: x.mode().iloc[0],
                                                    'DiabetesPedigreeFunction': 'mean'})
```

```python
aggregated_df
```

# **Perulangan dan Pemilihan Bersyarat**

## **Nested Loop**

```python
# Menggunakan nested loop untuk mencetak pola bintang segitiga
tinggi = 5

# Loop untuk baris
for i in range(tinggi):
    # Loop untuk kolom
    for j in range(i + 1):
        print("*", end=" ")
    print()  # Pindah ke baris baru setelah mencetak setiap baris
```

## **Nested If**

```python
# Meminta pengguna untuk memasukkan nilai
nilai = float(input("Masukkan nilai Anda: "))

# Menggunakan nested if untuk menentukan grade
if nilai >= 90:
    if nilai >= 95:
        print("Grade: A+")
    else:
        print("Grade: A")
elif nilai >= 80:
    print("Grade: B")
elif nilai >= 70:
    print("Grade: C")
elif nilai >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

## **Loop x Conditions**

```python
# Inisialisasi variabel total_pembelian
total_pembelian = 0

# Daftar barang beserta harganya
daftar_barang = {
    "Apel": 2.00,
    "Pisang": 1.00,
    "Jeruk": 1.50,
    "Melon": 3.00,
    "Anggur": 4.00
}

# Menampilkan daftar barang
print("Daftar Barang:")
for barang, harga in daftar_barang.items():
    print(f"{barang}: ${harga:.2f}")

# Meminta pengguna untuk memasukkan barang yang dibeli
while True:
    barang_dibeli = input("Masukkan barang yang dibeli (ketik 'selesai' untuk selesai): ").title()
    if barang_dibeli == 'Selesai':
        break

    # Memeriksa apakah barang tersedia dalam daftar
    elif barang_dibeli in daftar_barang:
        total_pembelian += daftar_barang[barang_dibeli]
    else:
        print("Maaf, barang tidak tersedia.")

# Menampilkan total pembelian
print(f"Total pembelian: ${total_pembelian:.2f}")
```
